```python
# Intializing librarys
import os
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, callbacks
from sklearn.utils.class_weight import compute_class_weight
from sklearn.metrics import classification_report, confusion_matrix
import matplotlib.pyplot as plt
import itertools
import cv2

# Mounting Google drive for the dataset
from google.colab import drive
drive.mount('/content/drive')
ZIP_PATH = '/content/drive/MyDrive/nndataset.zip'
EXTRACT_TO = '/content/skin_dataset'
DATASET_ROOT = os.path.join(EXTRACT_TO, 'IMG_CLASSES')
MODEL_SAVE_PATH = '/content/drive/MyDrive/skin_resnet50_safe.h5'

#Intializing the parameters
IMG_SIZE = 224
BATCH = 16
SEED = 123
EPOCHS_BASE = 8
EPOCHS_FINETUNE = 6
UNFREEZE_LAST_N = 20
USE_MIXED_PRECISION = False



if USE_MIXED_PRECISION:
    try:
        from tensorflow.keras import mixed_precision
        mixed_precision.set_global_policy('mixed_float16')
        print('Mixed precision enabled')
    except Exception as e:
        print('Could not enable mixed precision:', e)
        USE_MIXED_PRECISION = False
else:

    try:
        from tensorflow.keras import mixed_precision
        mixed_precision.set_global_policy('float32')
    except Exception:
        pass


try:
    from google.colab import drive
    drive.mount('/content/drive')
except Exception:
    print('Not running in Colab or google.colab not available. Proceeding without Drive mount.')

#Unzip the dataset
if not os.path.exists(EXTRACT_TO):
    if os.path.exists(ZIP_PATH):
        print('Extracting dataset from zip...')
        os.makedirs(EXTRACT_TO, exist_ok=True)
        !unzip -q "{ZIP_PATH}" -d "{EXTRACT_TO}"
    else:
        raise FileNotFoundError(f'Zip not found at {ZIP_PATH}. Update ZIP_PATH or upload dataset.')
else:
    print('Dataset folder already exists:', EXTRACT_TO)


if not os.path.exists(DATASET_ROOT):
    raise FileNotFoundError(f"Dataset root not found at {DATASET_ROOT}. Check EXTRACT_TO and folder structure.")


# Dataset: use image_dataset_from_directory
AUTOTUNE = tf.data.AUTOTUNE

print('Creating training and validation datasets...')
train_ds = tf.keras.utils.image_dataset_from_directory(
    DATASET_ROOT,
    validation_split=0.30,
    subset='training',
    seed=SEED,
    image_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH
)

val_ds = tf.keras.utils.image_dataset_from_directory(
    DATASET_ROOT,
    validation_split=0.30,
    subset='validation',
    seed=SEED,
    image_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH
)

class_names = train_ds.class_names
NUM_CLASSES = len(class_names)
print('Detected classes:', NUM_CLASSES, class_names)

# Preprocessing  augmentation
preprocess_input = tf.keras.applications.resnet.preprocess_input

data_augmentation = keras.Sequential([
    layers.RandomFlip('horizontal'),
    layers.RandomRotation(0.08),

], name='simple_augmentation')

def prepare_dataset(ds, training=False):
    def _map_fn(x, y):
        x = tf.cast(x, tf.float32)
        if training:
            x = data_augmentation(x)
        x = preprocess_input(x)
        return x, y

    ds = ds.map(_map_fn, num_parallel_calls=AUTOTUNE)
    ds = ds.prefetch(AUTOTUNE)
    return ds

train_ds = prepare_dataset(train_ds, training=True)
val_ds = prepare_dataset(val_ds, training=False)

# Compute class weights  for class imbalnce
print('Computing class weights from folder counts (fast and memory-safe)...')
counts = []
labels_list = []
for idx, cls in enumerate(class_names):
    cls_folder = os.path.join(DATASET_ROOT, cls)
    files = [f for f in os.listdir(cls_folder) if f.lower().endswith(('.png','.jpg','.jpeg','.bmp','.tif','.tiff'))]
    n = len(files)
    counts.append(n)
    labels_list.extend([idx] * n)

counts = np.array(counts)
print('Per-class image counts:', dict(zip(class_names, counts.tolist())))

# compute class weights using sklearn
classes_unique = np.unique(labels_list)
class_weights_raw = compute_class_weight('balanced', classes=classes_unique, y=np.array(labels_list))
# map to {class_index: weight}
class_weights = {int(i): float(w) for i, w in enumerate(class_weights_raw)}
print('Class weights:', class_weights)

# Build ResNet50 model (feature-extraction)
print('Building ResNet50 model (feature-extraction stage)...')
base_model = tf.keras.applications.ResNet50(
    input_shape=(IMG_SIZE, IMG_SIZE, 3),
    include_top=False,
    weights='imagenet'
)
base_model.trainable = False

inputs = keras.Input(shape=(IMG_SIZE, IMG_SIZE, 3))
x = base_model(inputs, training=False)
x = layers.GlobalAveragePooling2D()(x)
x = layers.Dropout(0.3)(x)

if USE_MIXED_PRECISION:
    outputs = layers.Dense(NUM_CLASSES, activation='softmax', dtype='float32')(x)
else:
    outputs = layers.Dense(NUM_CLASSES, activation='softmax')(x)

model = keras.Model(inputs, outputs)

opt = keras.optimizers.Adam(learning_rate=5e-4)
if USE_MIXED_PRECISION:
    from tensorflow.keras import mixed_precision
    opt = mixed_precision.LossScaleOptimizer(opt)

model.compile(optimizer=opt,
              loss=keras.losses.SparseCategoricalCrossentropy(),
              metrics=['accuracy'])

model.summary()

checkpoint_cb = callbacks.ModelCheckpoint(MODEL_SAVE_PATH, save_best_only=True, monitor='val_accuracy', mode='max')
reduce_lr = callbacks.ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=3, verbose=1)
early_stopping = callbacks.EarlyStopping(monitor='val_loss', patience=6, restore_best_weights=True)


# Train - feature extraction
print('Starting feature-extraction training...')
history = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=EPOCHS_BASE,
    callbacks=[checkpoint_cb, reduce_lr, early_stopping],
    class_weight=class_weights
)

# Fine-tune (unfreeze last N layers)
print('Preparing fine-tuning: unfreezing last', UNFREEZE_LAST_N, 'layers of ResNet50')
base_model.trainable = True
if UNFREEZE_LAST_N > 0:
    total_layers = len(base_model.layers)
    freeze_until = max(0, total_layers - UNFREEZE_LAST_N)
    for i, layer in enumerate(base_model.layers):
        layer.trainable = True if i >= freeze_until else False
else:
    for layer in base_model.layers:
        layer.trainable = False
opt_fine = keras.optimizers.Adam(learning_rate=1e-5)
if USE_MIXED_PRECISION:
    from tensorflow.keras import mixed_precision
    opt_fine = mixed_precision.LossScaleOptimizer(opt_fine)

model.compile(optimizer=opt_fine,
              loss=keras.losses.SparseCategoricalCrossentropy(),
              metrics=['accuracy'])

print('Starting fine-tuning...')
history_finetune = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=EPOCHS_FINETUNE,
    callbacks=[checkpoint_cb, reduce_lr, early_stopping],
    class_weight=class_weights
)

# Load best model if checkpoint saved
if os.path.exists(MODEL_SAVE_PATH):
    print('Loading best model from', MODEL_SAVE_PATH)
    model = keras.models.load_model(MODEL_SAVE_PATH)
else:
    print('No checkpoint found; using current model object.')

# -----------------------
# Evaluation: classification report and confusion matrix
# -----------------------
print('Running evaluation on validation set (building y_true/y_pred)...')
y_true = []
y_pred = []
for images, labels in val_ds:
    preds = model.predict(images)
    y_true.extend(labels.numpy().tolist())
    y_pred.extend(np.argmax(preds, axis=1).tolist())

print('Classification Report:')
print(classification_report(y_true, y_pred, target_names=class_names))

cm = confusion_matrix(y_true, y_pred)
print('Confusion matrix shape:', cm.shape)

plt.figure(figsize=(10,8))
plt.imshow(cm, interpolation='nearest', cmap=plt.cm.Blues)
plt.title('Confusion matrix')
plt.colorbar()
tick_marks = np.arange(len(class_names))
plt.xticks(tick_marks, class_names, rotation=45, ha='right')
plt.yticks(tick_marks, class_names)
thresh = cm.max() / 2.
for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
    plt.text(j, i, format(cm[i, j], 'd'),
             horizontalalignment='center',
             color='white' if cm[i, j] > thresh else 'black')
plt.tight_layout()
plt.ylabel('True label')
plt.xlabel('Predicted label')
plt.show()


# -----------------------
# Save model (again)
# -----------------------
try:
    model.save(MODEL_SAVE_PATH)
    print('Saved final model to', MODEL_SAVE_PATH)
except Exception as e:
    print('Could not save model:', e)

print('Done.')
```

```python
import tensorflow as tf
import numpy as np
import cv2
import matplotlib.pyplot as plt

def get_img_array(img_path, size=224):
    img = tf.keras.utils.load_img(img_path, target_size=(size, size))
    array = tf.keras.utils.img_to_array(img)
    array = np.expand_dims(array, axis=0)
    array = tf.keras.applications.resnet.preprocess_input(array)
    return array

model = tf.keras.models.load_model('/content/drive/MyDrive/skin_resnet50_safe.h5')
base_model = model.get_layer('resnet50')
sample_img = "/content/test.jpg"
img_array = get_img_array(sample_img, size=224)

last_conv_layer = base_model.get_layer('conv5_block3_out')

gap_layer = model.get_layer('global_average_pooling2d')
dropout_layer = model.get_layer('dropout')
dense_layer = model.get_layer('dense')

with tf.GradientTape() as tape:
    conv_outputs = last_conv_layer(base_model(img_array, training=False), training=False)
    tape.watch(conv_outputs)
    x = gap_layer(conv_outputs)
    x = dropout_layer(x, training=False)
    preds = dense_layer(x)
    pred_index = tf.argmax(preds[0])
    class_channel = preds[:, pred_index]

# Compute gradients
grads = tape.gradient(class_channel, conv_outputs)
pooled_grads = tf.reduce_mean(grads, axis=(0, 1, 2))
conv_outputs = conv_outputs[0]  # shape: (7, 7, 2048)
heatmap = conv_outputs @ pooled_grads[..., tf.newaxis]
heatmap = tf.squeeze(heatmap)
heatmap = np.maximum(heatmap, 0) / (tf.reduce_max(heatmap) + 1e-8)
heatmap = heatmap.numpy()

def superimpose_heatmap(img_path, heatmap, alpha=0.4):
    img = cv2.imread(img_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    heatmap_resized = cv2.resize(heatmap, (img.shape[1], img.shape[0]))
    heatmap_colored = cv2.applyColorMap(np.uint8(255 * heatmap_resized), cv2.COLORMAP_JET)
    superimposed_img = cv2.addWeighted(img, 1-alpha, heatmap_colored, alpha, 0)
    return superimposed_img

superimposed_img = superimpose_heatmap(sample_img, heatmap, alpha=0.4)

plt.figure(figsize=(8,4))
plt.subplot(1,2,1)
plt.title("Original")
plt.imshow(tf.keras.utils.load_img(sample_img))
plt.axis("off")
plt.subplot(1,2,2)
plt.title("Grad-CAM")
plt.imshow(superimposed_img)
plt.axis("off")
plt.show()
```

```python
class_names = [
    'Eczema',
    'Warts Molluscum and other Viral Infections',
    'Melanoma',
    'Atopic Dermatitis',
    'Basal Cell Carcinoma (BCC)',
    'Melanocytic Nevi (NV)',
    'Benign Keratosis-like Lesions (BKL)',
    'Psoriasis pictures Lichen Planus and related diseases',
    'Seborrheic Keratoses and other Benign Tumors',
    'Tinea Ringworm Candidiasis and other Fungal Infections'
]

# Load and preprocess an image
img_path = "/content/test.jpg"
img_array = get_img_array(img_path, size=224)

# Get prediction
pred = model.predict(img_array)
pred_class_idx = np.argmax(pred)
pred_class = class_names[pred_class_idx]
pred_prob = pred[0][pred_class_idx]

print("Class probabilities:", pred[0])
print(f"Predicted class: {pred_class} (index = {pred_class_idx})")
print(f"Confidence: {pred_prob:.2%}")
```
